# Grouping by Gender and Fraud Status to sum the Claim Amounts

explanation :
copy_df.groupby(['PatientGender', 'ClaimLegitimacy']):
groupby: This groups the data in copy_df based on the values in the specified columns: PatientGender and ClaimLegitimacy.
PatientGender: It groups the data separately for each gender (e.g., Male and Female).
ClaimLegitimacy: For each gender, the data is further split by fraud status, which could be 0 (fraudulent) or 1 (legitimate).
So, this will create four groups:

Male - Fraudulent
Male - Legitimate
Female - Fraudulent
Female - Legitimate
2. ['ClaimAmount'].sum():
After grouping the data, we are specifying which column's values we want to aggregate. Here, its ClaimAmount.
.sum(): This computes the sum of the ClaimAmount for each group. For example, for all males who submitted legitimate claims, the total claim amount will be summed up, and similarly for each other group.
3. .reset_index():
By default, groupby operations return a DataFrame where the grouping columns (PatientGender and ClaimLegitimacy) are treated as an index.
reset_index(): This resets the index so that PatientGender and ClaimLegitimacy become regular columns in the resulting DataFrame. Without this step, they would be part of the index, which can make further operations more difficult.
Result:
The result is a new DataFrame grouped_data that looks like this:

PatientGender	ClaimLegitimacy	ClaimAmount
Female	0	total fraud claims by females
Female	1	total legitimate claims by females
Male	0	total fraud claims by males
Male	1	total legitimate claims by males

```python

# Grouping by Gender and Fraud Status to sum the Claim Amounts
grouped_data = copy_df.groupby(['PatientGender', 'ClaimLegitimacy'])['ClaimAmount'].sum().reset_index()

# Plotting
plt.figure(figsize=(10, 6))

# Barplot to show Total Claim Amount vs Fraud for Male and Female
sns.barplot(x='PatientGender', y='ClaimAmount', hue='ClaimLegitimacy', data=grouped_data)

# Labels and Title
plt.title('Total Claim Amount vs Fraud Status by Gender', fontsize=16)
plt.xlabel('Gender', fontsize=14)
plt.ylabel('Total Claim Amount', fontsize=14)

# Show plot
plt.show()


```

# Let's show the Maximum fraud_reported Vs occupation

```python
# Filter data for fraudulent claims (ClaimLegitimacy == 0)
fraud_data = copy_df[df['ClaimLegitimacy'] == 0]

# Group by occupation (PatientEmploymentStatus) and find the maximum fraud amount
grouped_fraud = fraud_data.groupby('PatientEmploymentStatus')['ClaimAmount'].max().reset_index()

# Sort values by ClaimAmount to make it more readable (optional)
grouped_fraud = grouped_fraud.sort_values(by='ClaimAmount', ascending=False)

# Plot the results
plt.figure(figsize=(12, 8))
sns.barplot(x='PatientEmploymentStatus', y='ClaimAmount', data=grouped_fraud, palette="viridis")

plt.title('Maximum Fraud Reported (Claim Amount) vs Occupation', fontsize=16)
plt.xlabel('Occupation', fontsize=14)
plt.ylabel('Maximum Fraud Claim Amount', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()

plt.show()
```

# maximum insured claim amount based on the education level(no education in our d ataset)

```python
# Group by Education Level and find the maximum Claim Amount

grouped_education = copy_df.groupby('PatientEducationLevel')['ClaimAmount'].max().reset_index()

# Sort values by ClaimAmount to make it more readable (optional)
grouped_education = grouped_education.sort_values(by='ClaimAmount', ascending=False)

# Plot the results
plt.figure(figsize=(12, 8))
sns.barplot(x='PatientEducationLevel', y='ClaimAmount', data=grouped_education, palette="coolwarm")

plt.title('Maximum Insured Claim Amount vs Education Level', fontsize=16)
plt.xlabel('Education Level', fontsize=14)
plt.ylabel('Maximum Insured Claim Amount', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()

plt.show()
```

## Group by ProviderLocation and count the number of incidents

```python
# Group by ProviderLocation and count the number of incidents
grouped_location = copy_df.groupby('ProviderLocation').size().reset_index(name='IncidentCount')

# Sort the values to show the location with the most incidents
grouped_location = grouped_location.sort_values(by='IncidentCount', ascending=False)

# Plot the results
plt.figure(figsize=(12, 8))
sns.barplot(x='ProviderLocation', y='IncidentCount', data=grouped_location, palette="viridis")

plt.title('Number of Incidents by Provider Location', fontsize=16)
plt.xlabel('Provider Location', fontsize=14)
plt.ylabel('Number of Incidents', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()

plt.show()
```

```python
# Grouping by Gender and Fraud Status to sum the Claim Amounts
grouped_data = copy_df.groupby(['PatientMaritalStatus', 'ClaimLegitimacy'])['ClaimAmount'].sum().reset_index()

# Plotting
plt.figure(figsize=(10, 6))

# Barplot to show Total Claim Amount vs Fraud for Male and Female
sns.barplot(x='PatientMaritalStatus', y='ClaimAmount', hue='ClaimLegitimacy', data=grouped_data)

# Labels and Title
plt.title('Total Claim Amount vs Fraud Status by Gender', fontsize=16)
plt.xlabel('Gender', fontsize=14)
plt.ylabel('Total Claim Amount', fontsize=14)

# Show plot
plt.show()
```

# Fradulent activity by Age Range

```python
# Filter only fraudulent claims (ClaimLegitimacy = 0 for fraudulent)
fraudulent_claims = df[df['ClaimLegitimacy'] == 0]

# Define age ranges (bins) for grouping
age_bins = [0, 20, 40, 60, 80, 100]
age_labels = ['0-20', '21-40', '41-60', '61-80', '81-100']

# Create a new column 'AgeRange' based on the bins

fraudulent_claims['AgeRange'] = pd.cut(fraudulent_claims['PatientAge'], bins=age_bins, labels=age_labels)

# Group by AgeRange and count the number of fraudulent claims
age_fraud_count = fraudulent_claims.groupby('AgeRange').size().reset_index(name='FraudCount')

# Plot the results
plt.figure(figsize=(10, 6))
sns.barplot(x='AgeRange', y='FraudCount', data=age_fraud_count, palette='coolwarm')

plt.title('Fraudulent Activity by Age Range', fontsize=16)
plt.xlabel('Age Range', fontsize=14)
plt.ylabel('Number of Fraudulent Claims', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.show()
```
